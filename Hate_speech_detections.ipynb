```python
data['labels'] = data['class'].map({0:"Hate Speech", 1 : "Offensive Language", 2: "No hate or offensive language"})
```

```python
data_1 = data[["tweet","labels"]]
```

```python
import re
import nltk
import string
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords = set(stopwords.words("english"))
```

```python
stemmer = nltk.SnowballStemmer("english")
```

```python
def clean_data(text):
  text = str(text).lower()
  text = re.sub('https?://\S+|www\.S+','',text)
  text = re.sub('\[.*?\]','',text)
  text = re.sub('<.*?>*','',text)
  text = re.sub('[%s]' %re.escape(string.punctuation) , '',text)
  text = re.sub('\n','',text)
  text = re.sub('\w"\d\w"','',text)
  text = [word for word in text.split(' ') if word not in stopwords]
  text = " ".join(text)
  text = [stemmer.stem(word) for word in text.split(' ') ]
  text = " ".join(text)
  return text
```

```python
data_1["tweet"] = data_1["tweet"].apply(clean_data)
```

```python
data_1
```

```python
x = np.array(data_1['tweet'])
y = np.array(data_1['labels'])
```

```python
x
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv = CountVectorizer()
x = cv.fit_transform(x)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cn = confusion_matrix(y_test , y_pred)
cn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.heatmap(cn, annot = True, fmt = ".1f", cmap = "YlGnBu")
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
